# 读取数据

```python
def unique_value(df: pd.DataFrame) -> np.array:
    df_specific = df.drop(list(df.columns[:2]), axis=1, inplace=False)
    temp = df_specific.to_numpy()
    return np.unique(temp)
```

## 训练集特征稀疏性可视化分析 + 食材分布可视化分析

```python
food_index = {}
food_count = {} # key为类, value为对应类的出现次数
loc_index = {} # key为类, value为其所有数据的索引
for _, group in train_df.groupby(by='cuisine'):
    # name, group = specific data -> DataFrame
    name = group['cuisine'].head(1).item()
    group.drop(["id", 'cuisine'], axis=1, inplace=True)
    loc_index[name] = group.index
    food_index[name] = group.sum().to_dict()
    food_count[name] = len(group)
food_index.keys() # key为类, value为类中各元素总和 dict: (column:sum_value)
```

```python
cuisine_count = list(food_index.keys())
plt.figure(figsize=(10, 25), dpi=250)
for i in range(len(cuisine_count)):
    cuisine = list(food_index.keys())[i]
    plt.subplot(len(cuisine_count), 1, i+1)
    # 对所有地方的美食的使用材料总数进行可视化
    plt.bar(range(len(food_index[cuisine].keys())), food_index[cuisine].values())
    plt.title(cuisine)

# 防止重叠
plt.tight_layout()
# 绘制
plt.savefig("./Ingredient analysis/ingredients_analysis_total.jpg")
# plt.show()
plt.close()
```

```python
# 找出标签数量top3的location, 之后进行以上画图操作即可
top3_location = [item[0] for item in sorted(food_count.items(), key=lambda item : item[1], reverse=True)[:3]]
```

```python
j = 1
for loc in top3_location:
    plt.figure(figsize=(10, 5), dpi=200)
    plt.bar(range(len(food_index[loc].keys())), food_index[loc].values())
    plt.title(loc)
    plt.xlabel('ingredients')
    plt.ylabel('sum_values')
    plt.savefig("./Ingredient analysis/ingredients_analysis_{loc}_top{j}.jpg".format(loc=loc, j=j))
    j += 1
    plt.close()
```

## 标签与特征的相关性可视化分析

```python
from sklearn.decomposition import PCA
features_names = train_df.columns[2::]
labels = train_df.loc[:, 'cuisine']
```

### 原始特征之间相关程度

```python
for cuisine in cuisine_count:
    # 只取有值的feature
    cur_features_names = [item[0] for item in food_index[cuisine].items() if item[1]!=0]
    cuisine_features = train_df.loc[loc_index[cuisine], cur_features_names]
    corr_matrix = cuisine_features.corr()
    plt.figure(figsize=(10, 10), dpi=100)
    sns.heatmap(corr_matrix, cmap='coolwarm', square=True, fmt='.2f')
    plt.title('{} Features Correlation Map'.format(cuisine.title()))
    plt.savefig('./Features Correlation Map/{} Features Correlation Map'.format(cuisine.title()))
    plt.close()
```

### PCA


```python
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(train_df.loc[:, features_names])
plt.figure(figsize=(10, 10), dpi=100)
sns.scatterplot(x=reduced_features[:, 0], y=reduced_features[:, 1], hue=labels)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.savefig("./PCA/2 Principal Components.jpg")
plt.close()
```

#### 主成分分析获得的主成分与原始特征（前100）的关联程度

```python
# 绘制热力图
components_coefficients = pca.components_
plt.figure(figsize=(20, 10), dpi=200)
sns.heatmap(components_coefficients[:, :100], cmap='coolwarm', \
    xticklabels=features_names[:100], yticklabels=['Principal Component 1', 'Principal Component 2'])
plt.xlabel('Original Features(first 100)')
plt.ylabel('Components')
plt.title('Correlation between Principal Components and Original Features(first 100)')
plt.savefig("./PCA/Heatmap Correlation between Principal Components and Original Features(first 100).jpg")
plt.close()
```
